```python
import os

import torch
from tqdm import tqdm
import pandas as pd
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from models.watermark_faster import watermark_model
```

```python
cd '../instances'
```

```python
model = watermark_model(language='English', mode=None, tau_word=0.8, lamda=0.83)
```

```python
def watermark_detect(model, raw, mode='precise'):
    is_watermark, p_value, n, ones, z_value = model.watermark_detector_fast(raw) if mode == 'fast' else model.watermark_detector_precise(raw)
    return {"z_score": z_value, "p_value": p_value, "prediction": is_watermark}

def calculate_scores(filename, keyword, custom_filename=None):
    detection_filename = (filename[:-5] + '_detection.json') if custom_filename == None else custom_filename
    df = pd.read_json(filename, lines=True)
    detected = 0
    not_detected = 0
    with open(detection_filename, "r+" if os.path.exists(detection_filename) else "w+") as f:
        with torch.no_grad():
            for i in tqdm(range(len(f.readlines()), len(df["input"]))):
                line = (df["input"][i] if keyword == "continuation" else "") + df[keyword][i]

                result = watermark_detect(model, line)

                _line = df["input"][i].replace('\n', '\\n').replace('"', '\\"')

                f.write(f'{{"input": "{_line}", "p_value": "{result["p_value"]}", "z_score": "{result["z_score"]}", "prediction": "{result["prediction"]}"}}\n')

                if result["prediction"]:
                    detected += 1
                else:
                    not_detected += 1

    print("\nDetected:", detected)
    print("Not detected:", not_detected)
```

```python
calculate_scores('llama_yang_no_attack.json', "continuation")
```

```python
calculate_scores('opt_yang_no_attack.json', "continuation")
```
