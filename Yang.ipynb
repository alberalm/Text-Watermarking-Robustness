```python
import os

import torch
from tqdm import tqdm
import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from demo_CLI import watermark_embed_demo
from models.watermark_faster import watermark_model
```

```python
cd '../instances'
```

```python
checkpoint = 'facebook/opt-1.3b'

model = AutoModelForCausalLM.from_pretrained(checkpoint, device_map="auto")

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
model.eval()

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
watermark = watermark_model(language='English', mode=None, tau_word=0.8, lamda=0.83)
```

```python
import string
p = set(string.printable)
```

```python
filename_1 = 'opt_no_watermark.json'
filename_2 = 'opt_yang_no_attack.json'
filename_3 = 'human_yang_no_attack.json'

df = pd.read_json('c4_selection.json', lines=True)

with open(filename_1, "r+" if os.path.exists(filename_1) else "w+") as f:
    with open(filename_2, "a+" if os.path.exists(filename_2) else "w+") as f_w:
        with open(filename_3, "a+" if os.path.exists(filename_3) else "w+") as f_w_og:
            with torch.no_grad():
                for i in tqdm(range(len(f.readlines()), len(df["input"]))):
                    line = df["input"][i]
                    tokenized_input = tokenizer(line, return_tensors='pt').input_ids.to(model.device)

                    output_tokens = model.generate(tokenized_input, max_new_tokens=250, num_beams=2, do_sample=True)

                    # if decoder only model, then we need to isolate the
                    # newly generated tokens as only those are watermarked, the input/prompt is not

                    output_text = tokenizer.decode(output_tokens[0][51:], skip_special_tokens=True)

                    _line = ''.join(filter(lambda x: x in p, line)).replace('\n', '\\n').replace('"', '\\"')
                    _output_text = ''.join(filter(lambda x: x in p, output_text)).replace('\n', '\\n').replace('"', '\\"')

                    f.write(f'{{"input": "{_line}", "continuation": "{_output_text}"}}\n')

                    w_output = watermark.embed(output_text)

                    _w_output_text = ''.join(filter(lambda x: x in p, w_output)).replace('\n', '\\n').replace('"', '\\"')

                    f_w.write(f'{{"input": "{_line}", "continuation": "{_w_output_text}"}}\n')

                    w_og_output = watermark.embed(df["continuation"][i])

                    _w_og_output_text = ''.join(filter(lambda x: x in p, w_og_output)).replace('\n', '\\n').replace('"', '\\"')

                    f_w_og.write(f'{{"input": "{_line}", "continuation": "{_w_og_output_text}"}}\n')
```

```python
filename = 'llama_yang_no_attack.json'

df = pd.read_json('llama_no_watermark.json', lines=True)

with open(filename, "r+" if os.path.exists(filename) else "w+") as f:
    with torch.no_grad():
        for i in tqdm(range(len(f.readlines()), len(df["input"]))):

            line = df["input"][i]
            llama_text = df["continuation"][i]

            output_text = watermark.embed(llama_text)

            _line = ''.join(filter(lambda x: x in p, line)).replace('\n', '\\n').replace('"', '\\"')
            _output_text = ''.join(filter(lambda x: x in p, output_text)).replace('\n', '\\n').replace('"', '\\"')

            f.write(f'{{"input": "{_line}", "continuation": "{_output_text}"}}\n')
```
